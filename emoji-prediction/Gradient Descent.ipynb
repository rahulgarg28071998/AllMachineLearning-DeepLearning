```python
import numpy as np # Mathematical Library
import matplotlib.pyplot as plt #For Graph Plotting
```

```python
x = np.arange(10)
print(x)
```

```python
def loss(x):
    return (x-4)**2

l = loss(x)
```

```python
print(l)
```

```python
plt.style.use("seaborn")
plt.plot(x,l)
plt.show()
```

```python
x_ = 0.1
lr = 0.2
plt.plot(x,l)

for i in range(30):
    plt.scatter(x_,loss(x_))
    x_ = x_ - lr*2*(x_-4)
    print("X %.4f Loss %.4f"%(x_,loss(x_)))
    
    
plt.show()
    
```
